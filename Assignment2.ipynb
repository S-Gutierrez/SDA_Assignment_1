We decided to follow the Professor's advice and picked Burglary and Fraud as our two categories. As we will be using weather data to better our model, one of the categories is highly dependent on the weather (Burglary) whereas the other not at all (Fraud). For this problem we want to predict a crime given a date, time and location without giving importance to the frequency of occurence. We have a predominanance of one class over other which would bias our model. Therefore we decide to balance the data and consider equal splits of both categories. 
If we decided not to balance our data our train split might be highly in-balanced (a drastic example is 80% class 1 to 20% class 2) and at that point our baseline would be of 80%, the case in which class 1 is always predicted. Our model would move towards always predicting class 1, missing some of the underlying patterns in the data.  

```python
import pandas as pd
import sklearn.preprocessing as preprocessing
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, plot_confusion_matrix
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource, FactorRange, Legend
from bokeh.plotting import figure
from bokeh import palettes
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv", header='infer')
df['Timestamp'] = df[['Date', 'Time']].agg('-'.join, axis=1)
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format="%m/%d/%Y-%H:%M")

df_crimes = df[df["Category"].isin(["BURGLARY", "FRAUD"])].reset_index(drop=True)
# Take equal number of samples for each category
df_crimes = pd.concat([df_crimes[df_crimes['Category'].isin(['BURGLARY'])].sample(10000),
                     df_crimes[df_crimes['Category'].isin(['FRAUD'])].sample(10000)])
```

As model input we utilize every temporal feature and PdDistrict as geological data. Crimes can be distributed differently inside the same neighbourhood but this information can also be redundant. Therefore we decided to only utilize the district as a location, in order to find city patterns while keeping the model more simple.


```python
# Create timestamp for a better management of date and time information
df_crimes["HourOfDay"] = df_crimes["Timestamp"].dt.hour
df_crimes["DayOfWeek"] = df_crimes["Timestamp"].dt.dayofweek
df_crimes["MonthOfYear"] = df_crimes["Timestamp"].dt.month
df_crimes = df_crimes[["Category", "PdDistrict", "DayOfWeek", "HourOfDay", "MonthOfYear"]]
```

We encode PdDistrict, as it is categorical data, using one-hot encoding in order to be able to handle the data. 
For the temporal data, on the other hand, we chose to encode it with sine and cosine, as the cyclical characteristics of the data are mantained with this kind of encoding.

```python
#encode the data
# Ordinal encoding of categorical input variables
df_crimes["Category"]=preprocessing.LabelEncoder().fit_transform(df_crimes["Category"])
df_crimes["DayOfWeek"]=preprocessing.LabelEncoder().fit_transform(df_crimes["DayOfWeek"])
# One hot encoding of categorical features
categorical=list(set(df_crimes.columns) - set(df_crimes._get_numeric_data().columns))
df_encoded = pd.get_dummies(df_crimes[categorical])
                        
df_encoded = pd.concat([df_crimes, df_encoded], axis=1)

df_encoded.drop(categorical, axis=1, inplace=True)
# sin/cos encode cyclical data
df_encoded['day_sin'] = np.sin(df_encoded["DayOfWeek"]*(2.*np.pi/7))
df_encoded['day_cos'] = np.cos(df_encoded["DayOfWeek"]*(2.*np.pi/7))
df_encoded['hr_sin'] = np.sin(df_encoded["HourOfDay"]*(2.*np.pi/24))
df_encoded['hr_cos'] = np.cos(df_encoded["HourOfDay"]*(2.*np.pi/24))
df_encoded['mnth_sin'] = np.sin((df_encoded["MonthOfYear"]-1)*(2.*np.pi/12))
df_encoded['mnth_cos'] = np.cos((df_encoded["MonthOfYear"]-1)*(2.*np.pi/12))
df_encoded.drop(["HourOfDay","MonthOfYear"], axis=1, inplace=True)

# label encode to the output variable
df_encoded["Category"]=preprocessing.LabelEncoder().fit_transform(df_encoded["Category"])
```

As we have enough data we can consider splitting our dataset into train-set and test-set. We train a RandomForest Classifier on 75% of the data and test on the rest of the data which was held out during training. This gives a good idea of how well our model will predict unseen data.

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    df_encoded.iloc[:,1:], df_encoded.iloc[:,0], test_size = 0.25, random_state = 42)
```

Finally, we train a randomforest with the training data and predict with the test data.

```python
# Instantiate model with 1000 decision trees
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(X_train, y_train);
```

```python
# Use the forest's predict method on the test data
y_pred = rf.predict(X_test)
```

We set a baseline with random selection. Another baseline we could use is accuracy of 50% as that is how our classes are split in our dataset

```python
# Set a baseline accuracy to beat
random_classifier = np.random.choice(y_test.unique(), len(y_test))
baseline_accuracy = np.sum(random_classifier == y_test) / len(y_test)
print("Baseline accuracy: %1.4f" %baseline_accuracy )
```

```python
# Verify accuracy of the random forest classifier
rf_accuracy = np.sum(y_pred == y_test) / len(y_test)
print("Random Forest accuracy: %1.4f" %rf_accuracy )
```

For a deeper analysis of the predictions we plot a normalized Confusion Matrix

```python
baseline_conf_matrix = confusion_matrix(y_test, random_classifier, normalize='true')

class_names = ['Fraud', 'Burglary']

#baseline
conf_baseline = ConfusionMatrixDisplay(confusion_matrix=baseline_conf_matrix,
                              display_labels=class_names)
conf_baseline = conf_baseline.plot( cmap=plt.cm.Reds)
conf_baseline.ax_.set_title("Baseline Confusion Matrix")
plt.show()

#RandomForest
conf_RandomForest = plot_confusion_matrix(rf, X_test, y_test,display_labels=class_names, cmap=plt.cm.Blues, normalize='true')
conf_RandomForest.ax_.set_title("RandomForest Confusion Matrix")
plt.show()
```

As we can see our model learn from the crimes patterns as it can be seen from the improvement with regard to the baseline. Our true values in the confusion matrix (categories correctly predicted) are consistent with the accuracy evaluated and it seems like the model makes mistakes equally between the two categories (in purple) so we can verify that our model is not biased to one of the two categories.

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
df = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv", header='infer')
df['Timestamp'] = df[['Date', 'Time']].agg('-'.join, axis=1)
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format="%m/%d/%Y-%H:%M")

df_crimes=df[(df['Timestamp'].dt.year>2010)&
             (df['Timestamp'].dt.year<2018)&
             (df['Category'].isin(focuscrimes))]
```

We need to classify in what hour of the day each of the focused crimes happened so before handing the dataframe to Bokeh, we reorganize it pivoting the Category column. The pivoted dataframe shows the relative count of crimes for each hour of the day in every crime category.

```python
df_crimes.loc[:,'HourOfTheDay']=df_crimes.loc[:,'Timestamp'].dt.hour

df_grouped = pd.pivot_table(df_crimes, index = "HourOfTheDay", 
                            columns = "Category",values = 'PdId' ,aggfunc = 'count')
df_grouped = df_grouped.div(df_grouped.sum(axis=0), axis=1)

# Create Hour of the day column as the pivot_table Pandas function converts the HourOfTheDay column to an index.
# We need Hour of the day as a separate column to be handed to Bokeh.
df_grouped['Hours']=df_grouped.index.values
```

Now, it is time to display the results in a more visual way in Bookeh. First, we convert the Pandas dataframe to Bokeh dataframe.

```python
output_notebook() # open the bokeh viz on the notebook.
## it is a standard way to convert your df to bokeh
source = ColumnDataSource(df_grouped)
```

The next step is to create an empty Bokeh figure to add the plotting data to it.

```python
p = figure(x_range=FactorRange(factors=list(map(str, (df_grouped.index+1).values.tolist()))), 
           plot_height=400, plot_width=800, title='Hourly Crimes Distribution',
           x_axis_label='Hour of the Day', y_axis_label='Proportioned Frequency')
```

We make use of bvar function, present inside bokeh library, to add each crime category bar chart to the figure created previously. We se the "muted" data almost transparent so it is only shown in the graph when picked. Additionally, the bars are also a bit transparent so that we can visualize several of them at the same time. To finish with a better visualization, the labels where placed at tbe left of the graph so it does not cover part of the graphs and the font size of the labels are shrinked so it can be fully visualized in the notebook.

```python
color = palettes.Category20[len(focuscrimes)]
bar ={} # to store vbars
items = []
### here we will do a for loop:
for indx,i in enumerate(focuscrimes):
    bar[i] = p.vbar(x='Hours', 
                    top=i, 
                    source=source, 
                    width=0.9,
                    muted=True, 
                    muted_alpha=0.005,
                    color=color[indx],
                    fill_alpha=0.7,
                    line_alpha=0.7)
    items.append((i, [bar[i]]))
    
legend = Legend(items=items)
p.add_layout(legend, 'left')
p.legend.click_policy = 'mute'
p.legend.label_text_font_size='7pt'
show(p)
```
